# Сравнение результатов тестов на нормальность

Целью работы является проверка таких тестов, как Шапиро-Вилка, Колмагорова-Смирнова и Д'агостина на идентичность результатов проверки распределения на нормальность. 



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import math
import random
from scipy import stats
```

Для простоты написания кода напишем следующие функции:

1) Функция "create_result" необходима для вывода результата. Выводит True, если p-value больше 0.05 и False, если меньше. В каждом тесте, в случае, если p-value больше 0.05, распределение считается нормальным. В противном случае оно не нормально. 

Входной параметр - list_1 является списком из p-value. 


2) Функция формирования выборки из 1000 случайных значений.

Входные параметры:

value - количество случайных значений (у нас оно будет = 1000)

lambda_v - параметр распределения сл.вел. (0,0125)

n - объем выборки (то есть количество раз выбора случайного числа)

Эту функцию я взял тут https://github.com/AlexanderPetrenko83/Articles/blob/master/20191011_Central%20limit%20theorem/CPT_2.ipynb

```python
def create_result(list_1):
  list_result = []
  for i in list_1:
    list_result.append(i > 0.05)
  return list_result

def sample_expon(value, lambda_v, n):
    mu = 1/lambda_v
    list_values = []
    for i in range(value):
      summa = 0
      for j in range(n):
        summa += random.expovariate(mu)
      list_values.append(summa/n)
    return list_values
```

Для получения результатов необходимо задать параметры функции "sample_expon", а так же вывести гистограмму распределений для наглядности. Количество раз выбора случайного числа примем от 1 до 30000 с разным шагом.  

```python
# Задаем параметры

lambda_v = 0.0125
value = 1000
n_values = [1, 2, 3, 5, 7, 10, 15, 50, 100, 250, 500, 1000, 2000, 3000, 5000, 10000, 30000]
```

```python
# Списки для сохранения результатов тестов

shapiro_wilka = []
kolmogorov_smirnov = []
d_agostinos = []

# Проводим итерацию по заданным в списке числам, обозначающим кол-во раз выбора случайного числа
for n in n_values:
      list_x = sample_expon(value, lambda_v, n)
      print(' Результат теста Шапиро-Уилка:', stats.shapiro(list_x)[1], ',',
            'Нормальное распределение' if stats.shapiro(list_x)[1] > 0.05 else 'Ненормальное распределение',

            '\n', 'Результат теста Колмогорова-Смирнова:', stats.kstest(list_x, 'norm')[1], ',',
            'Нормальное распределение' if stats.kstest(list_x, 'norm')[1] > 0.05 else 'Ненормальное распределение', 

            '\n', 'Результат теста Д-Агостина:', stats.normaltest(list_x)[1], ',',
            'Нормальное распределение' if stats.normaltest(list_x)[1] > 0.05 else 'Ненормальное распределение')
      
# Добавляем результаты тестов в списки     
      shapiro_wilka.append(stats.shapiro(list_x)[1])
      kolmogorov_smirnov.append(stats.kstest(list_x, 'norm')[1])
      d_agostinos.append(stats.normaltest(list_x)[1])

# Создаем гистограммы распределений
      plt.figure(figsize = (10, 7))
      sns.histplot(list_x, kde = True, color = 'green')
      plt.show()
      print(42*'-', 5*'||', 42*'-')
```

Полученные результаты формуруем в датафреймы. Для наглядности используем тепловую карту

```python
# Применяем фукнцию create_result для лучшей интерпретации результата 

shapiro_wilka_result = create_result(shapiro_wilka)
kolmogorov_smirnov_result = create_result(kolmogorov_smirnov)
d_agostinos_result = create_result(d_agostinos)
```

```python
# Формируем датафреймы полученных результатов

shapiro_wilka = pd.DataFrame({'values':n_values,
                              'values_shapiro':shapiro_wilka, 
                              'result':shapiro_wilka_result})
kolmogorov_smirnov = pd.DataFrame({'values':n_values,
                                   'values_ks':kolmogorov_smirnov, 
                                   'result':kolmogorov_smirnov_result})
d_agostinos = pd.DataFrame({'values':n_values,
                            'values_d_agostinos':d_agostinos, 
                            'result':d_agostinos_result})

shapiro_wilka['result'] = shapiro_wilka['result'].replace({False:'Ненормальное распределение', 
                                                 True: 'Нормальное распределение'})
kolmogorov_smirnov['result'] = kolmogorov_smirnov['result'].replace({False:'Ненормальное распределение', 
                                                 True: 'Нормальное распределение'})
d_agostinos['result'] = d_agostinos['result'].replace({False:'Ненормальное распределение', 
                                                 True: 'Нормальное распределение'})
all_results = pd.DataFrame({'values':n_values, 
                            'shapiro_wilka_result':shapiro_wilka_result,
                            'kolmogorov_smirnov_result':kolmogorov_smirnov_result,
                            'd_agostinos_result':d_agostinos_result})
```

```python
all_r = all_results.copy()
all_r['shapiro_wilka_result'] = all_r['shapiro_wilka_result'].replace({False:0, 
                                                                       True:1})
all_r['kolmogorov_smirnov_result'] = all_r['kolmogorov_smirnov_result'].replace({False:0, 
                                                                       True:1})
all_r['d_agostinos_result'] = all_r['d_agostinos_result'].replace({False:0, 
                                                                       True:1})
```

```python
plt.figure(figsize = (12, 10))
sns.heatmap(all_r.drop(['values'], axis = 1), linewidths=.5, annot = True)
```

Таким образом, тест Шапиро-Вилка и Д'агостина показали одинаковые результаты, чего не скажешь о тесте Колмогорова-Смирнова.

Часто случается, что разные тесты дают разные результаты, так как за ними стоит разная математика. Поэтому, для каких-либо проверок, в том числе и для проверок на нормальность распределения, следует применть несколько разных тестов.
